Diseña un Python que consuma gran volumen de datos y lo muestre visualmente
(gráfico, dashboard…)

```python

import plotnine as p9
import pandas as pd
#lee el fichero csv
fichero = pd.read_csv("https://raw.githubusercontent.com/PabloPalomino/Hito1/main/world_country_and_usa_states_latitude_and_longitude_values.csv")

#data = fichero corresponde a indicar al metodo ggplot la ruta del archivo
(p9.ggplot(data=fichero,
           #recorre los datos pasados 
           mapping=p9.aes(x='latitude',
                          y='longitude'))
 #dibuja los puntos en el grafico, como parametro se le ha pasado el tamaño de los mismos y el color
    +p9.geom_point(alpha=0.3,color="red")
)
```
